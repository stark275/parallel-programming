## Installer ArgoCD sur K8s

```shell
kubectl create namespace argocd
kubectl apply -n argocd -f https://raw.githubusercontent.com/argoproj/argo-cd/stable/manifests/install.yaml

```

- Verifier que tous les status soient 1/1

```shell
microk8s kubectl get all -n argocd
```

- Rendre le service argocd-server accessible sur le réseau Local en changeant son type de `ClusterIp` en `NodePort`

```shell
microk8s kubectl patch svc argocd-server -n argocd -p '{"spec": {"type": "NodePort"}}'
```
- Verifier le port externe attribué

```shell
kubectl get svc -n argocd
```

- Recuperer le mot de passe admin

```shell
microk8s kubectl get secret argocd-initial-admin-secret  -n argocd -o yaml
```

Resultat

```yaml
apiVersion: v1
data:
  password: cExHVWVNSkpVSTRBWnpZag==
kind: Secret
metadata:
  creationTimestamp: "2024-06-04T15:19:02Z"
  name: argocd-initial-admin-secret
  namespace: argocd
  resourceVersion: "157767"
  uid: 73f7b330-0caa-40a2-a6a2-82e677ff8e79
type: Opaque
```

- Recuperer la valeur du champ `password` (cExHVWVNSkpVSTRBWnpZag==) et le decoder de la base

```shell
echo cExHVWVNSkpVSTRBWnpZag== | base64 --decode
```

Resultat:

```
pLGUeMJJUI4AZzYj
```
